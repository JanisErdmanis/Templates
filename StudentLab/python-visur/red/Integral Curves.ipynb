```python
from sympy import *
var('x y')
y = Function('y')(x)
```

```python
eq = exp(2*x) - 1 + y - y.diff(x)
eq
```

```python
f = dsolve(eq,y)
f.subs({y:0})
```

```python
solve(f.subs({y:0,x:1}))
```

```python
yeq = solve(f.subs({'C1':1}),y)
yeq
yeqf = lambdify(x,yeq[0],'numpy') #būtiska ir 0
```

```python
xlis = linspace(1,10,100)
plot(xlis,yeqf(xlis))
```

```python
yeq_list = [solve(f.subs({'C1':i*80}),y)[0] for i in range(-4,4)]
yeq_list
```

```python
yeqf_list = [lambdify(x,i,'numpy') for i in yeq_list]
```

```python
x_val  = linspace(-10,7,100)
for i in yeqf_list: plot(x_val,i(x_val))
xlim([0,10])
ylim([-5000,8000])
```
