```python
x = linspace(0,12,20)
y = linspace(0,12,20)
X1, Y1 = meshgrid(x,y)
f = lambda x,y: (cos(y) - y*cos(x))/(x*sin(y) + sin(x) -1)
DY = f(X1,Y1)
S = sqrt(1 + DY**2)
quiver(X1,Y1,1/S,DY/S)

streamplot(X1,Y1,1/S,DY/S)
```

### Diferenciālvinādojuma nointegrēšana

```python

def vectorfield(data,x): 
    
    y = data[0]
    y_d = f(x,y)#(cos(y) - y*cos(x))/(x*sin(y) + sin(x) -1)
    return [y_d]
def integr(y_init,x_min,x_max): 
    x_ = linspace(x_min,x_max,1000)
    res = odeint(vectorfield,[y_init],x_)
    plot(x_,res[:,0],linewidth=2)

integr(5.,3,6)
quiver(X1,Y1,1/S,DY/S)
```

```python
yint = linspace(5.2,6.5,4)
for y in yint:
    try:
        integr(y,2,11)
    except:
        pass
quiver(X1,Y1,1/S,DY/S)
xlim(0,12)
ylim(0,12)
```
