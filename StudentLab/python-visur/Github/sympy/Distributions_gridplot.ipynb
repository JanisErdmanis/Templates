```python
from sympy import *
from sympy.stats import *
```

```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
def plot_dist(ax, rv, urange, la):
    D = density(rv)
    f = D.args[1]
    u = np.linspace(urange[0], urange[1], 500)
    v = np.array([re(f.subs(D.args[0][0],i).evalf()) for i in u], dtype=np.float)
    m = np.amax(v)
    col = list(np.random.rand(3))

    ax.fill_between(u,np.zeros(u.shape),v, color=col+[0.5])
    ax.plot(u,v, color=col)
    ax.grid(True)
    ax.set_xlim(urange)
    ax.set_ylim([0.0, 1.1*m])
    ax.set_ylabel(la)
```

```python
DI = []
# Some commented out due to issues
DI.append([ Arcsin("x", 0, 1), (0.001,0.999), "Arcsin" ])  # Singular endpoints
DI.append([ Benini("x", 1,2,3), (3,10), "Benini" ])  # x > sigma must hold
DI.append([ Beta("x", 2,5), (0,1), "Beta" ])
DI.append([ BetaPrime("x", 2, 3), (0, 3.5), "BetaPrime" ])
DI.append([ Cauchy("x", 1, 0.5), (-4,4), "Cauchy" ])
DI.append([ Chi("x", 2), (0,4), "Chi" ])
DI.append([ Dagum("x", 2,1,1), (0,3), "Dagum" ])
#DI.append([ Erlang("x", 2,1), (0,3), "Erlang" ])
DI.append([ Exponential("x", 1), (0,5), "Exponential" ])
DI.append([ Gamma("x", 3,2), (0,20), "Gamma" ])
#DI.append([ Kumaraswamy("x", 2,2), (0,1), "Kumaraswamy" ])
DI.append([ Laplace("x", 1,2), (-10,10), "Laplace" ])
DI.append([ Logistic("x", 2,1), (-5,10), "Logistic" ])
DI.append([ LogNormal("x", 0, 1), (0.001,3), "LogNormal" ])  # LogNormal works only in limit x->0
DI.append([ Maxwell("x", 1), (0,5), "Maxwell" ])
DI.append([ Nakagami("x", 0.5, 1), (0,3), "Nakagami" ])
DI.append([ Normal("x", 0, 1), (-3,3), "Normal" ])
DI.append([ Pareto("x", 3, 1), (1,4), "Pareto" ])
#DI.append([ QuadraticU("x", 1,3), (0, 4), "U quadratic" ])
#DI.append([ RaisedCosine("x", 2, 3), (-2,6), "Raised Cosine" ])
DI.append([ Rayleigh("x", 3), (0,10), "Rayleigh" ])
DI.append([ StudentT("x", 1), (-5,5), "StudentT" ])
DI.append([ Triangular("x", 1,4,3), (0,6), "Triangular" ])
DI.append([ Uniform("x", -1,3), (-5,5), "Uniform" ])
#DI.append([ UniformSum("x", 2), (0,5), "UniformSum" ])  # Computation takes forever
#DI.append([ VonMises("x", pi,0.5), (0, 2*3.1415926535), "von Mises" ])
DI.append([ Weibull("x", 1,5), (0, 2.5), "Weibull" ])
DI.append([ WignerSemicircle("x", 2), (-4, 4), "WignerSemicircle" ])
```

```python
N = len(DI)

NC = 4
NR = ceil(N / (1.0*NC))

f = figure(figsize=(NC*4, NR*3))

for i, dist in enumerate(DI):
    ax = subplot(NR, NC, i+1)
    plot_dist(ax, *dist)
```
