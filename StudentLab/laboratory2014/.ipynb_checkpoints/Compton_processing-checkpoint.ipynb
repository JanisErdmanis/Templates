## The Compton effect

If the photon colides with electron and then changes the direction then Its momentum will decrease. It means that if we are observing colided photons then their wavelength will be larger due to:
$$p = \frac{h}{\lambda}$$
By amount of:
$$\Delta \lambda = \lambda_c (1 - \cos \phi)$$

The checking the validity of last equation is the purpose of this experiment

## Prepearing the environment

Firstly I am going to use two libaries from *pandas* - *read_table* and *concat*. The other need is to use vectorized functions which comes from *pylab*. The same libary also gives plotting routines.

```python
from pandas import read_table,concat
import pylab as plt

from IPython.display import Image # specific for notebook
```

The statement *from pylab import* and then dismissing all *plt* from code is faster for coding but it also makes it harder to learn from code. 

## Giving python the data

```python
table = read_table('table.dat') # imported from pandas

# Lets update namespace with new definitions
for col,series in table.iterkv():
    globals()[col]=series
```

Now the variables $angle,\, energy, \, FWHM$ have been created. Its time to compute new columns from them:


```python
# Just precise values in SI for various physical constants 
from scipy import constants

wavelength = constants.h*constants.c/(energy*constants.e) * 1e10 
error = FWHM/energy * wavelength

argument = 1 - plt.cos(angle*plt.pi/180) 
```

## Ploting

```python
fig = plt.figure() # good habit

plt.errorbar(argument,wavelength, yerr=error, fmt ='.')
plt.xlabel(r'$1-\cos \phi$')
plt.ylabel(r'$\Delta \lambda \, [pm]$')

fig.savefig('plot_data.png')
```

```python
Image('plot_data.png')
```

### Fitting the curve for the data

```python
# Definition of the curve where  m and n are parameters
f = lambda x,m,n:m*x + n

# Getting the best parameters with associated covariation matrix.
# sigma=error is optional but so powerfull!
# par are list [m,n]
from scipy.optimize import curve_fit
par, cov = curve_fit(f,argument,wavelength,sigma=error)
```

Now lets finish the figure by adding fitted curve. The used *figure(fig.number)* command tels that I want to continue this specific figure. Check what happens with other number!  

```python
fig_all = plt.figure(fig.number)

x = plt.linspace(0,2)
# but the *par just unpacks the list as direcly given f(x,n,m)
plt.plot(x,f(x,*par),label='par = {0} \n cov = {1}'.format(par,cov))
plt.legend(loc=2)

fig_all.savefig('plot_all.png')
```

```python
Image('plot_all.png')
```

## Creating the table for latex

The tables had been such a pain to make, that I almost returned to *gnumerics* (excell type program). Fortunately I found function *.to_latex* for table (more corectly the method of DataFrame) which directly makes the $\LaTeX$ formated table.

```python
# Asigning new names for columns
angle.name=r'$\phi \, [deg]$'
energy.name = r'$E \, [keV]$'
# FWHM has already satisfactory name
argument.name = r'$1 - \cos \phi$'
wavelength.name = r'$\lambda \, [pm]$'
error.name = '$\Delta \lambda$'

# Creating new DataFrame from series
# The wavelength.map('{:.3f}'.format) creates object with formated rows
# concat also comes from pandas
formated_table = concat([angle,energy,FWHM,argument,wavelength.map('{:.3f}'.format),error],axis=1)

# Output latex representation of the table
formated_table.to_latex('table.tex')
```
